```python
#find the mean orders for each department in the merged dataframe
#group by product name
ords_prods_merge.groupby('product_name')
```

```python
#group by department_id & calc mean by order number
ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

smaller subset is more variable and mean is lower. 
department w/ largest number orders is 'missing'

```python
#create loyalty flag
#group by user id, find & aggregate by max order
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#check
pd.options.display.max_rows = None
```

```python
ords_prods_merge.head(100)
```

```python
#flag
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#check frequency
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
#question 3 find the mean of product prices by loyalty group
#group by 'prices'
ords_prods_merge.groupby('prices')
```

```python
#split data into groups by "loyalty_flag" & calculate mean for each "prices"
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean']})
```

```python
#question 4
#group by user id, mean by product price/user, create agg as mean_price
ords_prods_merge['mean_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#check
ords_prods_merge.head(100)
```

```python
#flags for user id based on mean product price
ords_prods_merge.loc[ords_prods_merge['mean_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
#check freq of spending flag
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
#question 5 create order freq flag that indicates users' ordering behavior
#group data by "user_id" col, gen mean for days since prior order
ords_prods_merge['median_days'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#check
ords_prods_merge.head(100)
```

```python
#create flags w frequency labels to user id based on median days since prior order
ords_prods_merge.loc[ords_prods_merge['median_days'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days'] > 10) & (ords_prods_merge['median_days'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
#check freq via count
ords_prods_merge['frequency_flag'].value_counts(dropna=False)
```

```python
#print null values
print(ords_prods_merge[ords_prods_merge.frequency_flag.isnull()])
```

Null values are bc customer had one order so median cannot be calculated.
This person will be a non frequent customer

```python
#changing column heading
ords_prods_merge['frequency_flag'].fillna("Non-frequent customer", inplace = True)
```

```python
#check 
ords_prods_merge['frequency_flag'].value_counts(dropna=False)
```
